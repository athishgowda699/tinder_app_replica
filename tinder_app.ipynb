```python
import json,re
from datetime import datetime as dt

class tinder:
    def __init__(self):
        self.loginid=[]
        print('\n------** Welcome to Tinder **------')
        self.menu()
        
    def menu(self):
        while True:
            print('\n-----* Welcome to home page *-----')
            print('-------** Find your love **-------')
            print('\t1.Login')
            print('\t2.Registration')
            print('\t3.Exit') 
            try:    
                choice1=int(input('\n> Enter a choice : '))
                if choice1==1:
                    self.login()
                    break
                elif choice1==2:
                    self.registration()
                    break
                elif choice1==3:
                    break
                else:
                    print('\nEnter the proper number')
            except ValueError:
                print('\nEnter the proper key')


    def function(self):              
        while True:
            print('\n-----* Welcome to your profile page *-----')
            print('\t1.Look for the match')
            print('\t2.Your profile')
            print('\t3.Update profile')
            print('\t4.Logout')
            try:
                choice=int(input('\n > Enter a choice : '))
                if choice==1:
                    self.match()
                    break
                elif choice==2:
                    self.view_profile()
                    break                   
                elif choice==3:
                    self.update()
                    break
                elif choice==4:
                    self.loginid.pop()
                    self.menu()
                    break
                else:
                    print('\nEnter the proper number')
                    continue                    
            except ValueError:
                print('\nEnter the proper key')
                continue

    def login(self):
        c=input('\n> Enter your login id : ')
        d=input('> Enter the password : ')

        with open ("athish.json") as g:
            data1=json.load(g)
        l1=[]
        l2=[]
        for i1 in data1["users"]:
            l1.append(i1['id'])
            l2.append(i1['password'])
        if (c in l1 and d in l2):
            print('\nLogged in successfully......')
            self.loginid.append(c)
            self.function()
        else:
            print('\nEnter the correct details or do the registration if not done')
            self.menu()


    def view_profile(self):
        v=input('\n> Enter your login ID : ')
        if v in self.loginid:
            user=[]
            with open ("athish.json") as g:
                data1=json.load(g)

            for i2 in data1["users"]:
                if v==i2["id"]:
                    user.append(i2["details"])
                else:
                    pass

            print('\n\t>Name : ',user[0]['name'])
            print('\t>DOB : ',user[0]['dob'])
            print('\t>Mobile Number : ',user[0]['number'])
            print('\t>Email : ',user[0]['email'])
            print('\t>Tag line : ',user[0]['tag'])
            print('\t>About : ',user[0]['about'])
            print('\t>Gender : ',user[0]['gender'])
            print('\t>Preference : ',user[0]['preference'])
            print('\t>Interested in : ',user[0]['interest'])
            self.function()
        
        else:
            print('\nlogin id not found')
            self.function()


    def registration(self):

        l=[]
        with open ("athish12.json") as g:
            data1=json.load(g)
        for i3 in data1["users"]:
            l.extend([i3['id'],i3['password'],i3['details']['number'],i3['details']['email']])


        while True:
            login_id=input('> Enter the id you want to : ')

            if login_id not in l:
                break
            else:
                print('\nAlready in use...Use different login ID')
                continue

        while True:
            print('\nPassword should be of atleast one number')
            password=input('> Enter the password you want to : ')
            
            if re.search('[0-9]',password):            
                if password not in l:
                    break
                else:
                    print('\nAlready in use...Use different password')
                    continue
            else:
                print('\nEnter the proper password please')

        name=input('> Enter your full name : ').title()

        while True:
            number=input('> Enter number : ')
            
            if re.search('[0-9]{10}',number):
                if number not in l:
                    break
                else:
                    print('\nAlready in use...Use different number : ')
                    continue
                
            else:
                print('\nEnter the proper 10 digit mobile number')
                continue

        while True:
            email=input('> Enter email : ')
            if (re.search('[a-zA-Z]+',email)) and (re.search('@',email)) and (re.search('[.]',email)):
                if (email not in l):
                    break
                else:
                    print('\nAlready in use...Use different mail Id : ')
                    continue
            else:
                print("\nMention the proper mail ID\nI think you missed '@' or '.'")
                continue                

        while True:
            try:
                print('> Enter your DOB in date-month-year format : ')
                print('For eg. 10-04-1998')
                b1=input()
                dob1=dt.strptime(b1,'%d-%m-%Y')
                dob=dob1.strftime('%d-%m-%Y')
                break
            except:
                print('\nEnter the detail in proper format')
                continue
        tag=input('> Enter tagline : ').capitalize()
        about=input('> Enter about : ').capitalize()
        gender=input('> Enter gender : ').lower()
        
        print('''Enter your preference : 
Male,Female.
(male female)--format''')
        preference1=list(map(str,input().split(' ')))
        preference=[]
        for z in preference1:
            if z=='':
                pass
            else:
                preference.append(z)
        
        print('''\n> Please enter your interests :
Food, Cricket, Art, Reading, Travelling, Bikes, Cars.
(food cricket art)--format''')
        interest1=list(map(str,input().split(' ')))
        interest=[]
        for b in interest1:
            if b=='':
                pass
            else:
                interest.append(b)
            
        data={"id":login_id,"password":password,"details":{"name":name,"dob":dob,"number":number,"email":email,"tag":tag,"about":about,"gender":gender,"preference":preference,"interest":interest}}
               
        with open('athish.json') as f:
            info=json.load(f)
            temp=info["users"]
            temp.append(data)
        with open('athish12.json','w') as g:
            json.dump(info,g,indent=1)
        print('\nRegistration is successfully done.....')
        self.menu()
        

    def match(self):
        d=input('> Enter your login Id : ')
        print('\n')
        if d in self.loginid:
        
            with open ("athish.json") as g:
                data1=json.load(g)
            for i4 in data1["users"]:
                if d in i4["id"]:
                    l=1
                    v={}
                    for j in data1["users"]:           
                        if j["details"]["gender"] in i4["details"]["preference"]:
                            print(l,'Name : ',j["details"]["name"])
                            print(' ','Gender : ',j["details"]["gender"])
                            print(' ','Interests : ',j["details"]["interest"])
                            print('\n')
                            v[l]=[j["details"]["name"]]
                            l+=1
                    
                    if len(v)<=0:
                        print('No one matches for you to show...sorry ')
                    else:
                        print('\n> Select the number you want to have a Match with : ')
                        in1=list(map(int,input().split(' ')))
                        if len(v)<len(in1):
                            print('\nEnter the proper number or You might have selected more people than matching for you')
                        else:
                            list1=[]
                            for k in v:
                                if k in in1:
                                    list1.append(v[k])
                            if len(list1)<=0:
                                print('\nEnter the proper number please')
                            else:
                                print('\nMatch request have been sent to : ')
                                for i5 in list1:
                                    print(i5)

                else:
                    pass
            self.function()
        else:
            print('\nLogin Id not found')
            self.function()
            
    def update(self):

        l6=[]
        with open ("athish.json") as g:                       
                data3=json.load(g)
                data4=data3["users"]

        for i3 in data3["users"]:
            l6.extend([i3['password'],i3['details']['number'],i3['details']['email']])

        n1=input('\nEnter your login ID : ')
        if n1 in self.loginid:
            while True:
                try:
                    print('\n1.Password\n2.Name\n3.Number\n4.Email\n5.Tag\n6.About\n7.Preference\n8.Interest\n9.Go back')
                    print('\nEnter the number you want to edit/update : ')
                    in5=int(input())

                    if in5==1:       
                        while True:
                            print('\nPassword should be of atleast one number')
                            nwpass=input('> Enter the new password : ')
                            if re.search('[0-9]',nwpass):            
                                if nwpass not in l6:
                                    break
                                else:
                                    print('\nAlready in use...Use different password')
                                    continue
                            else:
                                print('\nEnter the proper password please')

                        for i5 in data4:
                            if i5['id']==n1:
                                i5['password']=nwpass  

                        with open('athish.json','w') as h:
                            json.dump(data3,h,indent=1)

                        print('\nPassword has been changed successfully')

                    elif in5==2:
                        nwname=input('> Enter the new name : ')

                        for i5 in data4:
                            if i5['id']==n1:
                                i5['details']['name']=nwname  

                        with open('athish.json','w') as h:
                            json.dump(data3,h,indent=1)

                        print('\nName has been changed successfully')


                    elif in5==3:
                        while True:
                            nwnum=input('> Enter new number : ')
                            if re.search('[0-9]{10}',nwnum):
                                if nwnum not in l6:
                                    break
                                else:
                                    print('\nAlready in use...Use different number : ')
                                    continue
                            else:
                                print('\nEnter the proper 10 digit mobile number')
                                continue

                        for i5 in data4:
                            if i5['id']==n1:
                                i5['details']['number']=nwnum 

                        with open('athish.json','w') as h:
                            json.dump(data3,h,indent=1) 

                        print('\nNumber has been changed successfully')


                    elif in5==4:       
                        while True:
                            newemail=input('> Enter email : ')
                            if (re.search('[a-zA-Z]+',newemail)) and (re.search('@',newemail)) and (re.search('[.]',newemail)):
                                if (newemail not in l6):
                                    break
                                else:
                                    print('\nAlready in use...Use different mail Id : ')
                                    continue
                            else:
                                print("\nMention the proper mail ID\nI think you missed '@' or '.'")
                                continue

                        for i5 in data4:
                            if i5['id']==n1:
                                i5['details']['email']=newemail  

                        with open('athish.json','w') as h:
                            json.dump(data3,h,indent=1)

                        print('\nEmail has been changed successfully')


                    elif in5==5:
                        nwtag=input('> Enter the new Tag : ')

                        for i5 in data4:
                            if i5['id']==n1:
                                i5['details']['tag']=nwtag 

                        with open('athish.json','w') as h:
                            json.dump(data3,h,indent=1)

                        print('\nTag has been changed successfully')


                    elif in5==6:
                        nwabout=input("> Enter the updated 'about' option : ")

                        for i5 in data4:
                            if i5['id']==n1:
                                i5['details']['about']=nwabout 

                        with open('athish.json','w') as h:
                            json.dump(data3,h,indent=1)

                        print('\nTag has been changed successfully')

                    elif in5==7:
                        print('''Enter your preference : 
Male,Female.
(male female)--format''')
                        nwpre1=list(map(str,input().split(' ')))
                        nwpre=[]
                        for z in nwpre1:
                            if z=='':
                                pass
                            else:
                                nwpre.append(z)

                        for i5 in data4:
                            if i5['id']==n1:
                                i5['details']['preference']=nwpre 

                        with open('athish.json','w') as h:
                            json.dump(data3,h,indent=1)

                        print('\nPreference has been changed successfully')

                    elif in5==8:
                        print('''\n> Please enter your interests :
        Food, Cricket, Art, Reading, Travelling, Bikes, Cars.
        (food cricket art)--format''')
                        nwint1=list(map(str,input().split(' ')))
                        nwint=[]
                        for b in nwint1:
                            if b=='':
                                pass
                            else:
                                nwint.append(b)

                        for i5 in data4:
                            if i5['id']==n1:
                                i5['details']['interest']=nwint 

                        with open('athish.json','w') as h:
                            json.dump(data3,h,indent=1)

                        print('\nInterests has been changed successfully')

                    elif in5==9:
                        self.function()
                        break

                    else:
                        print('\nEnter the proper number')

                except ValueError:
                    print('\nEnter the proper key')

        else:
            print('Login Id not found')
            self.function()
        
```
